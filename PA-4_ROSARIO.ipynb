## 2. Create a visualization that shows how the different features contributes to average grade. Does chosen track in college, gender, or hometown contributes to a higher average score



### *determining the average of each students*

```python
df['Average']=df.mean(axis=1, numeric_only=True)
df
```

### *Creating Plot for Visualization*

```python
features = ['Gender' , 'Track' ,'Hometown']
width = 0.6

plt.bar(df['Gender'] , df['Average'], width, align= 'edge')
plt.bar(df['Track'] , df['Average'], width, align= 'edge')
plt.bar(df['Hometown'] , df['Average'], width, align= 'edge')

plt.legend(['Gender', 'Track', 'Hometown'], loc='lower right')

plt.xlabel('Features')
plt.ylabel('Average')
plt.xticks(rotation = 48)

plt.show()
```
